```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hauling_historico_csv.csv', parse_dates=['time'], index_col='time')

df['month'] = df.index.month
df['day'] = df.index.day

df.head(5)
```

```python
df.kmh_plano.resample('d').mean().plot(color='g',figsize=(10,6))
df['kmh_-4'].resample('d').mean().plot(color='r',figsize=(10,6))
df['kmh_+4'].resample('d').mean().plot(color='b',figsize=(10,6))

plt.axhline(df.kmh_total.mean(), color='black', label='Promedio')
plt.xlabel("Tiempo")
plt.ylabel("Velocidad (km/h)")
plt.title("Velocidad Promedio por día")
plt.legend()

plt.show()
```

```python
df = df[df['dist_plano'] < 3000]
df['dist_plano'].resample('d').mean().plot(color='g',figsize=(10,6))
df['dist_-4'].resample('d').mean().plot(color='r',figsize=(10,6))
df['dist_+4'].resample('d').mean().plot(color='b',figsize=(10,6))

# plt.axhline(df.distance.mean(), color='black', label='Promedio')
plt.xlabel("Tiempo")
plt.ylabel("Distancia (m)")
plt.title("Distancia Promedio por día")
plt.legend()

plt.show()
```

```python
df_dia = pd.pivot_table(df, index=['año','month','day','tipo_ciclo'],
                       values=['kmh_+4','kmh_-4','kmh_plano','kmh_total'])
df_dia = df_dia.reset_index()
```

```python
df['month'] = df.index.month
df['day'] = df.index.day

df_mes = pd.pivot_table(df, index=['año','month','tipo_ciclo'],
                       values=['kmh_+4','kmh_-4','kmh_plano','kmh_total'])
df_mes = df_mes.reset_index()
```

```python
acarreando = df_dia[df_dia['tipo_ciclo'] == 'Acarreando']
viajando = df_dia[df_dia['tipo_ciclo'] == 'Viajando']

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot()

sns.distplot(acarreando['kmh_-4'], hist=False, ax=ax, label='A favor', color='g')
sns.distplot(acarreando['kmh_+4'], hist=False, ax=ax, label='En contra', color='r')
sns.distplot(acarreando['kmh_total'], hist=False, ax=ax, label='A nivel', color='b')

ax.set_title('Distribución de las Velocidades', fontsize=15, pad=15)
ax.set_xlabel('Velocidad (km/h)')
ax.set_ylabel('Densidad')
ax.legend(loc='best');
```

```python
avg_acarreando = acarreando['kmh_total'].mean()

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot()

sns.lineplot(x=acarreando.index,y=acarreando['kmh_-4'], ax=ax, label='A favor', color='g')
sns.lineplot(x=acarreando.index,y=acarreando['kmh_+4'], ax=ax, label='En contra', color='r')
sns.lineplot(x=acarreando.index,y=acarreando['kmh_plano'], ax=ax, label='A nivel', color='b')
sns.lineplot(x=acarreando.index,y=avg_acarreando, ax=ax, label='Promedio', color='black')
ax.set_title('Distribución de las Velocidades', fontsize=15, pad=15)
ax.set_xlabel('Tiempo')
ax.set_ylabel('Velocidad (km/h)')
ax.legend(loc='best');
```

```python
sns.set(style='whitegrid',palette='RdBu_r')
fig = plt.figure(figsize=(20,12))
axes = fig.subplots(2,3, sharex=True, sharey=True)

sns.boxplot(data=acarreando, x='año', y='kmh_plano', ax=axes[0,0])
sns.boxplot(data=acarreando, x='año', y='kmh_-4', ax=axes[0,1])
sns.boxplot(data=acarreando, x='año', y='kmh_+4', ax=axes[0,2])
sns.boxplot(data=viajando, x='año', y='kmh_plano', ax=axes[1,0])
sns.boxplot(data=viajando, x='año', y='kmh_-4', ax=axes[1,1])
sns.boxplot(data=viajando, x='año', y='kmh_+4', ax=axes[1,2])

axes[0, 0].set(xlabel='', ylabel='Cargado (km/h)', title='Velocidad a Nivel')
axes[0, 1].set(xlabel='', ylabel='', title='Velocidad a Favor')
axes[0, 2].set(xlabel='', ylabel='', title='Velocidad en Contra')
axes[1, 0].set(xlabel='', ylabel='Descargado (km/h)', title='')
axes[1, 1].set(xlabel='', ylabel='', title='')
axes[1, 2].set(xlabel='', ylabel='', title='')

fig.suptitle('Distribución de las velocidades (descartando valores atípicos)');
```
